4. Can keywords be used as variable names 
No, keywords cannot be used as variable names in Python.
Keywords are reserved words that have special meaning in Python’s syntax. Since Python already uses them to perform specific functions, they cannot be used as identifiers (variable names, function names, class names, etc.).

6. Why are lists mutable, but tuples are immutable 
List are mutable because lists are meant to:
Store collections that may change
Allow adding/removing elements
Support dynamic operations like:
append()
remove()
insert()
pop()
Lists are used when data needs to be modified

tuples are immutable because tuples are meant to:
Store fixed data
Be safe from accidental modification
Be used as dictionary keys
Be faster than lists

Conditional statements are used to make decisions in a program.

12. 4&8 How does the elif statement work 
It is used to check multiple conditions one by one after an if statement.
Python checks the if condition first.
If if is False → it checks the first elif.
If that is False → it checks the next elif.
When one condition becomes True → that block runs.
After that → Python stops checking further conditions.
If none are True → else block runs (if present).

🔹 Syntax
if condition1:
    # block 1
elif condition2:
    # block 2
elif condition3:
    # block 3
else:
    # block 4
🔹 Example
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
Output:
Grade B
✔ Because:
marks >= 90 → False
marks >= 70 → True
So it prints Grade B and stops.
It does NOT check the next conditions after that.
